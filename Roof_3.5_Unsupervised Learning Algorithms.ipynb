```python
from datetime import datetime
from IPython.display import display, Markdown
import pandas as pd
import numpy as np

todays_date = datetime.now().strftime("%Y-%m-%d")
student_id = "lacaug0782"

display(Markdown("## Step 1.2: Load the Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

file_path = r"C:\Users\lacey\OneDrive\Documents\ECPI 2024\SDC485 AI_ML Data Analytics Methods_Modeling\3.5PA\winequality-red.csv"
data = pd.read_csv(file_path, sep=",")

print(f"Data Shape: {data.shape}")
display(data.head())
```

```python
display(Markdown("## Step 1.3: Initial Exploration of the Dataset"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

print("Dataset Information:")
print(data.info())
print("\nSummary Statistics:")

display(data.describe())
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
display(Markdown("## Step 2.1: Preprocess the Data"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

print("Shape Before Cleaning:", data.shape)

data = data.drop_duplicates()

print("Shape After Removing Duplicates:", data.shape)
print("\nMissing Values After Cleaning:")
print(data.isnull().sum())
```

```python
display(Markdown("## Step 2.2: Standardize Numerical Features"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

from sklearn.preprocessing import StandardScaler

features = data.drop("quality", axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

print("Standardized data shape:", scaled_features.shape)
```

```python
display(Markdown("## Step 3.1: Hierarchical Clustering Dendrogram"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(scaled_features, method="ward")
plt.figure(figsize=(10, 7))
dendrogram(
    Z,
    truncate_mode="level",
    p=5,
    leaf_rotation=90,
    leaf_font_size=12,
    show_contracted=True
)

plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Sample Index / Cluster Size")
plt.ylabel("Distance")
plt.show()
```

```python
display(Markdown("## Step 4.1: Implement K-Means Clustering"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

from sklearn.cluster import KMeans

k = 3
kmeans = KMeans(
    n_clusters=k,
    random_state=42,
    n_init=10
)
kmeans.fit(scaled_features)

cluster_labels = kmeans.labels_

data["Cluster"] = cluster_labels

print("K-Means clustering complete.")
print("\nCluster Counts:")
print(data["Cluster"].value_counts())
```

```python
display(Markdown("## Step 5.1: Elbow Method for Optimal k"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

inertia = []

k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
    
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
display(Markdown("## Step 5.2: Silhouette Score for Optimal k"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

from sklearn.metrics import silhouette_score
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )
    labels = kmeans.fit_predict(scaled_features)
    score = silhouette_score(scaled_features, labels)
    silhouette_scores.append(score)
    
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Optimal k")
plt.show()

best_k = range(2, 11)[silhouette_scores.index(max(silhouette_scores))]

print("Best k based on silhouette score:", best_k)
print("Best silhouette score:", max(silhouette_scores))
```

```python
display(Markdown("## Step 6.1: Refit K-Means with Optimal k"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

optimal_k = best_k

final_kmeans = KMeans(
    n_clusters=optimal_k,
    random_state=42,
    n_init=10
)

final_labels = final_kmeans.fit_predict(scaled_features)

data["Final_Cluster"] = final_labels

final_silhouette = silhouette_score(
    scaled_features,
    final_labels
)

print("Optimal Number of Clusters:", optimal_k)
print("Final Silhouette Score:", final_silhouette)

print("\nFinal Cluster Counts:")
print(data["Final_Cluster"].value_counts())
```

```python
display(Markdown("## Step 6.2: Review Cluster Characteristics"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

cluster_summary = data.groupby("Final_Cluster").mean(numeric_only=True)
display(cluster_summary)
```

```python
display(Markdown("## Step 7.1: Visualize Clustering Results"))
display(Markdown(f"Date: {todays_date}"))
display(Markdown(f"Student ID: {student_id}"))

import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=data,
    x="alcohol",
    y="volatile acidity",
    hue="Final_Cluster",
    palette="Set2"
)
plt.title("Wine Clusters Based on Alcohol and Volatile Acidity")
plt.xlabel("Alcohol")
plt.ylabel("Volatile Acidity")
plt.legend(title="Cluster")
plt.show()
```
