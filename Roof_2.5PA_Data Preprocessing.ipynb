```python
from datetime import datetime
from IPython.display import display, Markdown
todays_date = datetime.now().strftime("%Y-%m-%d")
student_id = "lacaug0782"

display(Markdown("# Data Preprocessing on Breast Cancer Dataset"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

display(Markdown("## Step 2: Load and Explore Dataset"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

data = pd.read_csv("C:/Users/lacey/OneDrive/Documents/ECPI 2024/SDC485 AI_ML Data Analytics Methods_Modeling/2.5PA/breastCancerDataset.csv")
display(data.head())
print(data.info())
display(data.describe())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

display(Markdown("## Step 3: Preprocess the Data"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

# Check missing values
print(data.isnull().sum())

# Fill missing numerical values with median
data = data.fillna(data.median(numeric_only=True))

# Remove duplicates
data = data.drop_duplicates()

print(f"Dataset shape after removing duplicates: {data.shape}")

# Standardize selected features
scaler = StandardScaler()

features_to_scale = ['concavity1', 'concavity3']

data[features_to_scale] = scaler.fit_transform(
    data[features_to_scale]
)

# Convert diagnosis labels to numbers
data['Diagnosis'] = data['Diagnosis'].map({
    'B': 0,
    'M': 1
})

# Define features and target
X = data.drop('Diagnosis', axis=1)
y = data['Diagnosis']

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42
)

print(f"Training samples: {X_train.shape[0]}")
print(f"Testing samples: {X_test.shape[0]}")
```

```python
from sklearn.svm import SVC

display(Markdown("## Step 4: Implement SVM"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

svm_model = SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
display(Markdown("## Step 5: Evaluate SVM Model"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

svm_accuracy = accuracy_score(y_test, y_pred_svm)
print(f"Accuracy Score: {svm_accuracy}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.model_selection import GridSearchCV

display(Markdown("## Step 6: GridSearchCV for SVM"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

param_grid = {
   'C': [0.1, 1, 10, 100],
   'kernel': ['linear', 'rbf', 'poly']
}
grid_search_svm = GridSearchCV(
   SVC(),
   param_grid,
   refit=True,
   verbose=2,
   cv=5
)
grid_search_svm.fit(X_train, y_train)
print("Best Parameters:")
print(grid_search_svm.best_params_)
```

```python
display(Markdown("## Step 7: Evaluate Tuned SVM"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

best_svm = grid_search_svm.best_estimator_
y_pred_tuned_svm = best_svm.predict(X_test)
tuned_accuracy = accuracy_score(y_test, y_pred_tuned_svm)

print(f"Accuracy Score: {tuned_accuracy}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_tuned_svm))
```

```python
import xgboost as xgb

display(Markdown("## Step 8: Implement XGBoost"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

xgb_model = xgb.XGBClassifier(
use_label_encoder=False,
eval_metric='logloss'
)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
```

```python
display(Markdown("## Step 9: Evaluate XGBoost"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

xgb_accuracy = accuracy_score(y_test, y_pred_xgb)

print(f"Accuracy Score: {xgb_accuracy}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_xgb))
```

```python
display(Markdown("## Step 10: GridSearchCV for XGBoost"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

param_grid_xgb = {
'n_estimators': [50, 100],
'learning_rate': [0.1, 0.2],
'max_depth': [3, 4]
}
grid_search_xgb = GridSearchCV(
 xgb.XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss'
),
   param_grid_xgb,
   refit=True,
   verbose=2,
   cv=5
)
grid_search_xgb.fit(X_train, y_train)

print("Best Parameters:")
print(grid_search_xgb.best_params_)
```

```python
display(Markdown("## Step 11: Evaluate Tuned XGBoost"))
display(Markdown(f"### Date: {todays_date}"))
display(Markdown(f"### Student ID: {student_id}"))

best_xgb = grid_search_xgb.best_estimator_
y_pred_tuned_xgb = best_xgb.predict(X_test)
tuned_xgb_accuracy = accuracy_score(
    y_test,
    y_pred_tuned_xgb
)

print(f"Accuracy Score: {tuned_xgb_accuracy}")
print("\nClassification Report:")
print(classification_report(
y_test,
y_pred_tuned_xgb
))
```
